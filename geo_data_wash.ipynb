```python
import csv 
 
f = open('最新数据.csv', 'r')
csvreader = csv.reader(f)
final_list = list(csvreader)
```

```python
import pandas as pd 

df = pd.DataFrame(data=final_list[1:], columns=final_list[0])
```

```python
import json 

def read_json(file):
    f1 = open(file, 'rb')
    china = json.loads(f1.read().decode())
    items = []
    for pro in china['features']:
        p = pro['properties']
        item = {
            p['name']:{
            'long': p['cp'][0],
            'lat': p['cp'][1]
            }
        }
        items.append(item)
    print(items)
    return items
```

```python
import os 




provinces = ['./geometryProvince/'+each for each in os.listdir('./geometryProvince')]
countries = ['./geometryCouties/'+ each for each in os.listdir('./geometryCouties')]
china = 'china.json'

locations = []
locations.append(china)
locations.extend(provinces)
locations.extend(countries)
```

```python
ps = []
for file in locations:
    ps.extend(read_json(file))
```

```python
df['city_long'] = df['city_name']
df['city_lat'] = df['city_name']

df





```

```python
def get_long(x):
    for item in ps:
        for key, value in item.items():
            if x in key:
                print(item)
                return value['long']

df['city_long'] = df['city_long'].map(lambda x: get_long(x))
```

```python
def get_lat(x):
    for item in ps:
        for key, value in item.items():
            if x in key:
                print(item)
                return value['lat']


df['city_lat'] = df['city_lat'].map(lambda x: get_lat(x))
```

```python
df['经度'] = df.city_long
df['纬度'] = df.city_lat
```

```python
c = ['雇主id', '雇主名字', 'provice_name', 'city_name', '订单内容', '成交金额', '经度', '纬度']

new_df = df[c]
```
